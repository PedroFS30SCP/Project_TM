```python
# Imports
import pandas as pd
import numpy as np
import re
```

# Limpeza do Ficheiro imdb_reviews_test_ORIGIN.csv

> Ficheiro origial com as Reviwes de Teste IMDB

```python
# Carregamento do dataset de teste
df_test = pd.read_csv("../dataset/imdb_reviews_test_ORIGIN.csv") 
df_test.head()
```

```python
# Verificar o novo tamanho do dataset
print(f"Número de linhas do CSV: {len(df_test)}")
```

```python
# Verificar valores nulos
print(df_test.isnull().sum())

# Verificar número de linhas duplicadas
print(f"Número de linhas duplicadas: {df_test.duplicated().sum()}")

# Se quiser ver as linhas duplicadas
duplicadas = df_test[df_test.duplicated()]
print(duplicadas)
```

```python
# Agrupar os textos duplicados
duplicados_texto = df_test[df_test.duplicated(subset=["text"], keep=False)]

# Agrupar por texto e listar os índices onde aparecem
grupos = duplicados_texto.groupby('text').apply(lambda x: x.index.tolist())

# Inicializar um contador
contador_total = 0

# Imprimir duplicados e contar
for texto, indices in grupos.items():
    if len(indices) > 1:
        print(f"Texto duplicado encontrado {len(indices)} vezes nos índices: {indices}")
        print(f"Texto: {texto}")
        print("-" * 80)
        contador_total += (len(indices) - 1)  # Só contar as repetições extra

print(f"\nNúmero total de duplicações encontradas: {contador_total}")
```

```python
# Remover duplicados baseados apenas na coluna "text"
df_train = df_test.drop_duplicates(subset=["text"], keep='first').reset_index(drop=True)

# Verificar o novo tamanho do dataset
print(f"Número de linhas após remover duplicados: {len(df_train)}")
```

```python
def normalizar_espacos(text):
    return re.sub(r'\s+', ' ', text).strip()

# Percorrer todas as linhas
for idx, row in df_train.iterrows():
    texto_original = row['text']
    texto_corrigido = normalizar_espacos(texto_original)
    
    # Se o texto corrigido for diferente do original (houve espaços a mais)
    if texto_original != texto_corrigido:
        print(f"Linha {idx} com espaços a mais:")
        print("Original:", repr(texto_original))
        print("Corrigido:", repr(texto_corrigido))
        print("-" * 100)
```

### Dados limpos guardados em imdb_reviews_test.csv

```python
df_train.to_csv('../dataset/imdb_reviews_test.csv', index=False)
```
